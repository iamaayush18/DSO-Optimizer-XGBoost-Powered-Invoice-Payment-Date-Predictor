# BUCKETING:-

```python
# Adding a Bucket Column to Bucketize/Cap all the Predicted_Delay Values...
Final.loc[Final['Predicted_Delay']<0,'Bucket'] = "Negative"
Final.loc[Final['Predicted_Delay']<15,'Bucket'] = "0-15"
Final.loc[(Final['Predicted_Delay']>=15) & (Final['Predicted_Delay']<30),'Bucket'] = "15-30"
Final.loc[(Final['Predicted_Delay']>=30) & (Final['Predicted_Delay']<45),'Bucket'] = "30-45"
Final.loc[(Final['Predicted_Delay']>=45) & (Final['Predicted_Delay']<=60),'Bucket'] = "45-60"
Final.loc[(Final['Predicted_Delay']>=60) & (Final['Predicted_Delay']<=100),'Bucket'] = "60-100"
Final.loc[(Final['Predicted_Delay']>=100) & (Final['Predicted_Delay']<=150),'Bucket'] = "100-150"
Final.loc[Final['Predicted_Delay']>150,'Bucket'] = ">150"
```
